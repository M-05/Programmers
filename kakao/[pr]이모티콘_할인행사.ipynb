```python
# 할인율이 모두 동일할 때.
# 만 적용해서 실패

import numpy as np

def solution(users, emoticons):
    cnt_list = [] # 최대값
    # cnt = 0
    total_list = [] # 최소값
    # total = 0

    discount_dict = {40:0.6, 30:0.7,20:0.8,10:0.9}

    for d in discount_dict.keys():
        cnt = 0
        total = 0
        for u in users:
            # 할인율 비교 40 >= 40
            if d >= u[0]:
                # 9600.0 11200.0 # 금액 비교
                temp_total = sum(np.array(emoticons) * discount_dict[d])
                if temp_total >= u[1]:
                    cnt += 1
                else:
                    total += temp_total

        cnt_list.append(cnt)
        total_list.append(total)

    return [max(cnt_list), total_list[cnt_list.index(max(cnt_list))]]
```

```python
from itertools import product # 모든 조합 찾기

def solution(users, emoticons):
    rates = [10, 20, 30, 40]
    rate_dict = {10:0.9, 20:0.8, 30:0.7, 40:0.6}
    em_len = len(emoticons)
    answer = [0, 0]

    for rate in product(rates, repeat=em_len): # (10, 10)
        cnt = 0 # 할인 비율이 바뀔 때 마다 초기화
        total = 0 # 유저들의 총 금액
        for user in users:
            pay = 0 # 한 유저의 총 금액
            for i in range(em_len):
                if rate[i] >= user[0]: # 할인율 끼리 비교
                    pay += emoticons[i] * rate_dict[rate[i]]

            if pay >= user[1]: # 최대 금액 이상일 경우 초기화
                cnt +=1
                pay = 0 # 유저의 금액 초기화

            total += pay # 한 유저의 금액을 총합에 넣기.

        if cnt >= answer[0]: # 현재 가입한 수가 기존 가입자 이상일 경우
            if cnt == answer[0]: # 현재 가입자 수와 기존 가입자 수 같을 경우
                answer[1] = max(answer[1], total) # 두가지 방법 중 제일 높은 금액 고르기
            else:
                answer[1] = total
            answer[0] = cnt # 현재 가입자 수로 최신화.

    return answer
```

```python
# https://velog.io/@dh1010a/Python-%ED%94%84%EB%A1%9C%EA%B7%B8%EB%9E%98%EB%A8%B8%EC%8A%A4150368-%EC%9D%B4%EB%AA%A8%ED%8B%B0%EC%BD%98-%ED%95%A0%EC%9D%B8%ED%96%89%EC%82%AC

def solution(users, emoticons):
    answer = [0, 0]
    data = [10, 20, 30, 40]
    discount = []
    def dfs(tmp, d): # 모든 경우의 할인율 조합을 구함
        if d == len(tmp):
            discount.append(tmp[:])
            return
        else:
            for i in data:
                tmp[d] += i
                dfs(tmp, d+1)
                tmp[d] -= i
    dfs([0]*len(emoticons), 0)

    for disc in discount: # 만들어진 모든 조합을 하나씩 살펴봄
        cnt = 0
        get = 0
        for i in users:
            pay = 0
            for j in range(len(disc)):
                if i[0] <= disc[j]:
                    pay += emoticons[j] * (100 - disc[j])/100
                if pay >= i[1]:
                    break
            if pay >= i[1]: # 만약 유저의 제한금액 초과시 플러스 구매
                pay = 0
                cnt += 1
            get += pay
        if cnt >= answer[0]: # 현재 최대값을 넘어가면 갱신
            if cnt == answer[0]:
                answer[1] = max(answer[1], get)
            else:
                answer[1] = get
            answer[0] = cnt

    return answer
```

### [itertools](https://docs.python.org/ko/3/library/itertools.html)

- product('ABCD', repeat=2) `모든 경우의 수`  

>AA AB AC AD BA BB BC BD CA CB CC CD DA DB DC DD

- permutations('ABCD', 2)

>AB AC AD BA BC BD CA CB CD DA DB DC

- combinations('ABCD', 2)

>AB AC AD BC BD CD

```python
from itertools import product, combinations, permutations
```

```python
rate = [10, 20, 30, 40]
for i in combinations(rate, len(rate)):
    print(i)
```

```python
rate = [10, 20, 30, 40]
for i in permutations(rate, len(rate)):
    print(i)
```

```python
rate = [10, 20, 30, 40]

for i in product(rate, repeat=len(rate)):
    print(i)
```
