# [수들의 합 1789](https://www.acmicpc.net/problem/1789)

서로 다른 N개의 자연수의 합이 S라고 한다.   
S를 알 때, 자연수 N의 최댓값은 얼마일까

```python
#  S(1 ≤ S ≤ 4,294,967,295)
```

```python
s = int(input())

cnt = 0
start = 0

while True:
    cnt += 1
    start += cnt

    if start > s:
        break
print(cnt -1)
```

# [소인수분해 11653](https://www.acmicpc.net/problem/11653)

 integer factorization)는 1보다 큰 자연수를 소인수(소수인 인수)들만의 곱

```python
# 정수 N (1 ≤ N ≤ 10,000,000)
```

```python
n = int(input())

num = 2
answer = []
while n != 1:
    if n % num == 0:
        n //= num
        answer.append(num)
    else:
        num += 1
print(*answer, sep='\n')
```

# [약수들의 합 9506](https://www.acmicpc.net/problem/9506)

```python
# 입력의 마지막엔 -1
```

```python
n = 28
answer = []
num = n // 2

while num > 0:
    if n % num ==0:
        answer.append(num)
    num -= 1
if sum(answer) == n:
    answer.sort()
    print(f"{n} = ", end="")
    print(*answer, sep=" + ")
    # print(*sorted(answer), sep=" + ")
else:
    print(f"{n} is NOT perfect.")
```

```python
import sys

while True:
    n = int(input())
    if n == -1: # 입력의 마지막엔 -1
        sys.exit()
        # break
    answer = []
    num = n // 2

    while num > 0:
        if n % num ==0:
            answer.append(num)
        num -= 1
    if sum(answer) == n: # 약수들의 합이라면
        answer.sort()
        print(f"{n} = ", end="")
        print(*answer, sep=" + ")
        # print(*sorted(answer), sep=" + ")
    else: # 약수들의 합 아니라면
        print(f"{n} is NOT perfect.")
```

# [화성 수학 5355](https://www.acmicpc.net/problem/5355)

@는 3을 곱하고,  
%는 5를 더하며,  
 #는 7을 빼는 연산자

```python
print(round(8, ndigits=2))
print("{:.2f}".format(8))
print('%.2f'%8)
```

```python
import sys

with open('./bj.txt', 'w') as f:
    f.write('''3
3 @ %
10.4 # % @
8 #''')

sys.stdin = open('./bj.txt', 'r')

input_data = sys.stdin.readline
```

```python
operation = {'@':"*3","%":"+5","#":"-7"}
for _ in range(int(input_data())):
    data = input_data().split()
    answer = data[0]
    for d in data[1:]:
        if d == '@': # (10.4 +5 -7) * 3 이렇게 계산 하기 위해.
            answer = eval(answer)
            answer = str(answer)
        answer += operation[d]
    print('%.2f'%eval(answer))
    print("{:.2f}".format(eval(answer)))
```

```python
T = int(input())

for t in range(T):
    map = list(input().split())
    num = float(map[0])
    for x in map[1:]:
        if x == "@":
           num *= 3
        if x == "%":
            num += 5
        if x == "#":
            num -= 7
    print("%0.2f" %num)
```

# [그릇 7567](https://www.acmicpc.net/problem/7567)

처음 높이 5  
같은 방향 5  
다른 방향 10  

```python
# 문자열의 길이는 3이상 50 이하
```

```python
# ()()()))( # 80
string = input()

height = 10
for i in range(len(string)-1):
    if string[i] == string[i +1]:
        height += 5
    else:
        height += 10
print(height)
```

```python
dish = input()
h = 5*(len(dish)+1) # 5 * (4 +1)

for i in range(1, len(dish)):
    if dish[i] != dish[i-1]: # 서로가 같지 않다면
        h += 5

print(h)
```

```python
import sys

T = int(input())

for i in range(T):
    a, b = map(int, sys.stdin.readline().split())
    aa, bb = a, b

    while a % b != 0: # 서로 나눌 때 나머지가 0일때까지
        a, b = b, a % b

    print(aa*bb//b)
```
